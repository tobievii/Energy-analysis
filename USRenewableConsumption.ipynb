```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt

api_key = "4a60fcfe3a762ea4e4d9276f8944aa76"
```

```python
# Electric Power Sector Conventional Hydroelectric Net Summer Capacity, Annual
hydroelectric_url = "http://api.eia.gov/series/?api_key=" + api_key + "&series_id=STEO.HVEPCAP_US.A"
```

```python
print(requests.get(hydroelectric_url))
```

```python
response = requests.get(hydroelectric_url).json()
```

```python
print(json.dumps(response, indent=2, sort_keys=True))
```

```python
# Create lists to determine x and y axis for graphs
years = []
hydroelectric_consumption = []

# Loop through Hydroelectric data API to call data being measured
for x in response["series"][0]["data"]:
    years.append(x[0])
    hydroelectric_consumption.append(x[1])
    
#values = response["series"][0]["data"][1]
years  = years[::-1]
hydroelectric = hydroelectric_consumption[::-1]
```

```python
# Total Petroleum and Other Liquids Product Supplied, Annual
plt.figure(figsize=(20,10))

plt.scatter(years, hydroelectric)
plt.title("Electric Power Sector Conventional Hydroelectric Net Summer Capacity, Annual")
plt.ylabel("MegaWatts")
plt.xlabel("Years")
plt.grid(True)

plt.show()
```

```python
# Electric Power Sector Conventional Hydroelectric Net Summer Capacity, Annual

plt.figure(figsize=(20,10))
# Create Line Graph
plt.plot(years, hydroelectric)
plt.title("Electric Power Sector Conventional Hydroelectric Net Summer Capacity, Annual")
plt.ylabel("MegaWatts")
plt.xlabel("Years")
plt.grid(True)

plt.show()
```
