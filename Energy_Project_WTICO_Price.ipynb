```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#URL West Texas Intermediate Crude Oil Price, Annual
url_price = "http://api.eia.gov/series/?api_key=4a60fcfe3a762ea4e4d9276f8944aa76&series_id=STEO.WTIPUUS.A"
```

```python
price = requests.get(url_price).json()
print(json.dumps(price, indent=4, sort_keys=True))
```

```python
year = []
price_value = []
for x in price['series'][0]['data']:
    year.append(x[0])
    price_value.append(x[1])
```

```python

year  = year[::-1]
price_value = price_value[::-1]
plt.figure(figsize=(20,10))
plt.scatter(year, price_value)
```

```python
plt.figure(figsize=(20,10))
plt.plot(year, price_value)



plt.show()
```
